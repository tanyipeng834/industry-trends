## III. Methodology

- To study the course intake as a whole, we will consolidate datasets 1 & 2 mentioned in Section II into a single dataframe. It is made simple as both datasets have the same columns. Similarly, datasets 3 & 4 will be combined to form another dataframe to analyse the respective industry trends.

### Manipulation of Student Datasets

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#import uni intake datasetS
uni_intake_df = pd.read_csv('universities-intake-enrolment-and-graduates-by-course.csv', thousands=',')
uni_intake_df.head()
```

```python
#observe the uni DF
display(uni_intake_df.dtypes)
display(uni_intake_df.shape)
display(uni_intake_df.isnull().sum())
display(uni_intake_df['course'].unique())
```

```python
#add uni column to DF
uni_intake_df['tertiary type'] = uni_intake_df['course'].apply(lambda x: "University")
uni_intake_df.head()
```

```python
#import poly intake dataset
poly_intake_df = pd.read_csv('polytechnics-intake-enrolment-and-graduates-by-course.csv', thousands = ',')
poly_intake_df.head()
```

```python
#observe the poly DF
display(poly_intake_df.dtypes)
display(poly_intake_df.shape)
display(poly_intake_df.isnull().sum())
display(poly_intake_df['course'].unique())
```

```python
# add poly column to DF
poly_intake_df['tertiary type'] = uni_intake_df['course'].apply(lambda x: "Polytechnic")
poly_intake_df.head()
```

```python
# concatenate uni and poly DFs
combined_intake_df = pd.concat([uni_intake_df, poly_intake_df])
combined_intake_df.reset_index(drop=True)
```

```python
# create DF with total cohort only + from years 2010 - 2019 + only selected courses 

#filter to sex MF only
total_intake_df = combined_intake_df.loc[combined_intake_df['sex'] == 'MF']

# filter further to only years 2010 - 2019
mask = (total_intake_df['year'] > 2009) & (total_intake_df['year'] <= 2019)
total_intake_df = total_intake_df.loc[mask]

# filter to course information technology 
total_intake_df= total_intake_df[total_intake_df['course'] == 'Information Technology']
                                
total_intake_df.reset_index(drop=True)
```

```python
female_intake_df = combined_intake_df.loc[combined_intake_df['sex'] == 'F']

# filter further to only years 2010 - 2019
mask = (female_intake_df['year'] > 2009) & (female_intake_df['year'] <= 2019)
female_intake_df = female_intake_df.loc[mask]

# filter to course information technology 
female_intake_df= female_intake_df[female_intake_df['course'] == 'Information Technology']
                                
female_intake_df.reset_index(drop=True)
```

```python
#Getting data for proportion of females for each year of enrolment.
female_enrolment = female_intake_df['enrolment']
female_df = pd.DataFrame(female_enrolment)
female_df = female_df.reset_index(drop=True)

overall_enrolment = total_intake_df['enrolment']
overall_df = pd.DataFrame(overall_enrolment)
overall_df = overall_df.reset_index(drop=True)

proportion = (female_df / overall_df) * 100
proportion
```

```python
# final course cohort regardless of tertiary type

final_student_df = total_intake_df.groupby('year').sum().sum(
    level=['year']).reset_index()

final_student_df
```

### Manipulation of Employment Datasets

```python
#import employment change csv
employ_change_df = pd.read_csv('employment-change-by-industry-level3.csv')
employ_change_df.head()
```

```python
# create DF with from years 2010 - 2019 + only selected industry3

# filter employ_change_df to years 2010 - 2019
mask2 = (employ_change_df['year'] > 2009) & (employ_change_df['year'] <= 2019)
filtered_employ_change_year = employ_change_df.loc[mask2]

# filter to industry - it and other information services
filtered_employ_change_year = filtered_employ_change_year[filtered_employ_change_year['industry3'] 
                                                          == 'it and other information services']


filtered_employ_change_year.reset_index(drop=True)

# as this is a small DF where everything is visible at one glance, we don't observe the DF.
```

```python
#import employment recruiment resignation csv
employ_recruit_df = pd.read_csv('average-monthly-recruitment-resignation-rate-by-industry-level-3.csv')
employ_recruit_df.head()
```

```python
# create DF with from years 2010 - 2019 + only selected industry3

# filter employ_change_df to 2010 - 2019
mask3 = (employ_recruit_df['year'] > 2009) & (employ_recruit_df['year'] <= 2019)
filtered_employ_recruit_year = employ_recruit_df.loc[mask3]

# filter to industry - it and other information services + architectural and engineering services
filtered_employ_recruit_year = filtered_employ_recruit_year[filtered_employ_recruit_year['industry3'] 
                                                            == 'it and other information services']


filtered_employ_recruit_year.reset_index(drop=True)

# as this is a small DF where everything is visible at one glance, we don't observe the DF.
```

```python
# as DF filtered_employ_recruit_year & DF filtered_employ_change_year are the same, other than the percentage columns,
# we can merge the 2 DFs.

# add recruitment rate column to DF filtered_employ_change_year
filtered_employ_change_year['recruitment_rate'] = filtered_employ_recruit_year['recruitment_rate'].values

# add resignation rate column to DF filtered_employ_change_year
filtered_employ_change_year['resignation_rate'] = filtered_employ_recruit_year['resignation_rate'].values

# create final DF with only relevant columns
employ_trend_df = filtered_employ_change_year[['year','industry3','employment_change','recruitment_rate',
                                                     'resignation_rate']]

# rename column industry3
final_employ_trend_df = employ_trend_df.rename(columns={'industry3':'industry'})

final_employ_trend_df.reset_index(drop=True) 
```

### Visualisation of Data

```python
import matplotlib.pyplot as plt


intake_df = final_student_df[['year','intake']]


# Convert to float so the y-axis values on the right side will be in order.
def convert_float(x):
    return(float(x))

final_employ_trend_df['employment_change'] = final_employ_trend_df['employment_change'].apply(convert_float)
```

```python
# Graph of employment change and Student Intake

ax = intake_df.plot.bar(x='year', y='intake', rot=0, color = 'green', label = 'Intake')
plt.ylim([4000, 5500])
ax2 = ax.twinx()
ax2.plot( ax.get_xticks(), final_employ_trend_df['employment_change'].values, linestyle='-', marker='o', linewidth=2.0, label = 'employment change' )
plt.legend()
plt.title('Employment Change and Intake for IT' )
plt.show()
```

<p> There appears to be little to no correlation between these 2 variables. </p>

```python
graduates_df = final_student_df[['year','graduates']]
intake_df = final_student_df[['year','intake']]
enrol_df = final_student_df[['year','enrolment']]
```

```python
# Graph for Employment Change and Number of Graduates
ax = graduates_df.plot.bar(x='year', y='graduates', rot=0, color = 'green', label = 'Graduates')
plt.ylim([3500, 4800])
ax2 = ax.twinx()
ax2.plot( ax.get_xticks(), final_employ_trend_df['employment_change'].values, linestyle='-', marker='o', linewidth=2.0, label = 'employment change' )
plt.legend()
plt.title('Employment Change and Grads for IT' )
plt.show()
```

<p> There appears to be little to no correlation between these 2 variables. </p>

```python

def convert_float(x):
    return(float(x))
final_employ_trend_df['recruitment_rate'] = final_employ_trend_df['recruitment_rate'].apply(convert_float)
```

```python
# Graph for Recruitment Rate and Number of Graduates
ax = graduates_df.plot.bar(x='year', y='graduates', rot=0, color = 'red', label = 'Graduates')
plt.ylim([3500, 4600])

ax2 = ax.twinx()
ax2.plot( ax.get_xticks(), final_employ_trend_df['recruitment_rate'].values, linestyle='-', marker='o', linewidth=2.0, label = 'Recruitment' )
plt.legend(loc = 2)
plt.title('Recruitment Rate and Graduates for IT' )
plt.show()
```

<p> There appears to be little to no correlation between these 2 variables. </p>

```python
# Rows 0 to 9 belong to Uni, rows 10 to 19 belong to polytechnic

proportion['year'] = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
uni_proportion = proportion[ :10]
poly_proportion = proportion[ 10:]


# Visualising proportion of female enrolment for poly and uni.
plt.plot( 'year', 'enrolment', data=uni_proportion, marker='o', markerfacecolor='blue', markersize=6, color='skyblue', linewidth=2, label='University')
plt.plot( 'year', 'enrolment', data=poly_proportion, marker='o', color='olive', linewidth=2, label='Polytechnic')
plt.ylabel('Proportion of females (%)')
plt.xlabel('Year')
plt.title('Proportion of female enrolment per year')
plt.legend()
plt.show()
```

<p>There is an increasing trend for university from 2010 to 2015 and and a decrease from 2015 to 2017, before increasing from 2017 to 2019.
   For polytechnic, there has beem a decrease all the way from 2010 to 2019. </p>

## Conclusion
-In conclusion, it appears that there is not much correlation between the number of uni/poly grads and the increase in employment and recruitment rate for the IT industry and vice versa.
This could be do to certain factors or variables that were not accounted for, such as students not pursuing a career in the IT industry despite studying IT in uni or poly.

-There has been a decreasing trend in number of females in IT related poly courses and a slight increasing trend for number of females in university IT courses. The decrease could be due to changing preferences in choice of polytechnic courses. Perhaps more females are choosing other courses instead of IT, hence leading to a decrease in female in polytechnic IT courses.
